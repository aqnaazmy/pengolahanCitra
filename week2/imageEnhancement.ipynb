## Translasi

```python
import numpy as np
from skimage import io, color
from matplotlib import pyplot as plt
```

```python
def translate(img, x, y):
    height, width = img.shape
    # Create a zero matrix with the same shape as the input image
    res = np.zeros((height, width))
    
    # Handle translations that move the image completely out of view
    if y >= height or x >= width:
        return res  # Return empty result if the translation is too large
    
    # Copy the shifted part of the original image to the new location
    res[y:height, x:width] = img[0:height-y, 0:width-x]
    return res
```

```python
# Read the image and convert to grayscale
i1 = color.rgb2gray(io.imread('../assets/flower.jpg'))

# Apply translation
i2 = translate(i1, 15, 25)

# Plot the original and translated images
plt.subplot(1, 2, 1)
plt.imshow(i1, cmap='gray')
plt.title('Original')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(i2, cmap='gray')
plt.title("Translated")
plt.axis("off")

plt.show()
```
